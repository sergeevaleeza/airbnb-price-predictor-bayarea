```python
# Import libraries
import pandas as pd
import glob
import os
```

```python
# Set your directory (if needed)
csv_dir = "../data/simplified"
os.chdir(csv_dir)
```

```python
# Find and merge the files
files = glob.glob("*_neighbourhoods.csv")

combined = []

for f in files:
    # Keep "County" in region name
    region = f.replace("_neighbourhoods.csv", "").replace("_", " ")

    df = pd.read_csv(f)
    df["region"] = region

    combined.append(df)

# Combine all into one DataFrame
merged_df = pd.concat(combined, ignore_index=True)
```

```python
# Inspect the result
merged_df.head()
```

```python
# Save the combined file
merged_df.to_csv("all_neighbourhoods.csv", index=False)
```

```python
# Find all reviews CSV files
files = glob.glob("*_reviews.csv")

combined = []

for f in files:
    # Keep "County" in region name
    region = f.replace("_reviews.csv", "").replace("_", " ")

    df = pd.read_csv(f)
    df["region"] = region

    combined.append(df)

# Combine all review DataFrames
merged_reviews = pd.concat(combined, ignore_index=True)

# Optional: Move region column to front
cols = ["region"] + [col for col in merged_reviews.columns if col != "region"]
merged_reviews = merged_reviews[cols]
```

```python
# Preview first few rows
merged_reviews.head()
```

```python
# Save to file
merged_reviews.to_csv("all_reviews.csv", index=False)
```

```python
# Find all listings CSV files
files = glob.glob("*_listings.csv")

combined = []

for f in files:
    # Preserve "County" in region name
    region = f.replace("_listings.csv", "").replace("_", " ")

    df = pd.read_csv(f)
    df["region"] = region

    combined.append(df)

# Combine all DataFrames
merged_listings = pd.concat(combined, ignore_index=True)

# Move 'region' column to front (optional)
cols = ["region"] + [col for col in merged_listings.columns if col != "region"]
merged_listings = merged_listings[cols]
```

```python
# Preview result
merged_listings.head()
```

```python
# Save to file
merged_listings.to_csv("all_listings.csv", index=False)
```
